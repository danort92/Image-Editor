```python
import os
is_ok=False
#finchè il percorso non esiste continua a chiedere di inserire il percorso dell'immagine corretto
while is_ok==False:
    try:
        PATH=input("Inserisci il percorso della directory: ")
        
        #se il percorso non esiste continua a chiedere il percorso dell'immagine
        if not os.path.exists(PATH)==True:
            raise ValueError("\nIl percorso inserito non esiste.") 
            continue
        else:
            is_ok=True
            
            os.chdir(PATH)
            
            go_on="Sì"
            i=1
            while go_on!="Stop":
                is_go_on=False
                if go_on=="Sì":
                    i=1
                    while is_go_on==False:
                        try:
                            file_name=input("Inserisci il nome dell'immagine nel formato name.ext: ")
                            if not file_name in os.listdir(PATH):
                                raise ValueError(f"\nL'immagine {file_name} non esiste nella directory inserita")
                                continue
                            else:
                                is_go_on=True
                                root,extension=os.path.splitext(file_name)
                                print(f"File: {file_name}.\nEstensione: {extension}.")
                                from PIL import Image
                                with Image.open(PATH+"\\"+file_name) as image:
                                    width,height=image.size
                                    proportion=width/height
                                    print(f"Risoluzione: {width}X{height}.")
                        except ValueError as e:
                            print(e)


                if go_on in ["Sì","No"]:
                    #chiede inserimento di una nuova risoluzione. 
                    #se non rispetta il formato continua a chiedere il reinserimento.
                    #se si preme invio lascia la risoluzione inziale.   
                    import re
                    new_resolution=None
                    while new_resolution==None:
                        try:
                            new_resolution=input("Inserire nuova risoluzione in formato LARGHEZZAxLUNGHEZZA. Lasciare il campo vuoto se non si intende modificare la risoluzione: ")
                            if new_resolution=="":
                                new_width,new_height=width,height
                            elif all([item.isdigit() for item in re.split("x|X| ",new_resolution)])==False or len(re.split("x|X| ",new_resolution))!=2:
                                    new_resolution=None
                                    raise ValueError("\nFormato risoluzione non valido.")
                                    continue
                            else:
                                new_width,new_height=re.split("x|X| ",new_resolution)
                                new_width,new_height=int(new_width),int(new_height)
                        except ValueError as e:
                            print(e)
                    new_proportion=new_width/new_height
                    print(f"Nuova risoluzione: {new_width}X{new_height}.")



                    #se la proporzione tra larghezza e altezza è diversa da prima si chiede
                    #all'utente se desidera mantenere quella originale. Se sì si ricalcola l'altezza
                    if new_proportion!=proportion:
                        same_proportion="X"
                        while same_proportion=="X":
                            try:
                                same_proportion=input("Vuoi mantenere le proporzioni originali? [Sì/NO]. Premere Invio per NO: ")
                                if same_proportion!="Sì" and same_proportion!="No" and same_proportion!="":
                                    same_proportion="X"
                                    raise ValueError("\nRisposta non valida.")
                                    continue
                                elif same_proportion=="Sì":
                                    new_height=int(round(new_width/proportion,0))
                                    print(f"Nuova risoluzione: {new_width}X{new_height}.")
                                else:
                                    print(f"Nuova risoluzione: {new_width}X{new_height}.")
                            except ValueError as e:
                                print(e)

                    #chiede all'utente se vuole che venga salvata nuova immagine a colori o in scala di grigi
                    #visualizza direttamente nel notebook immagine, ne apre una nuova fuori e salva nuova immagine
                    #come nome_immagine_NEW nella stessa directory dell'immagine inziale

                    bnw="X"
                    while bnw=="X":
                        try:
                            bnw=input("Vuoi convertire in bianco e nero? [Sì/NO]: ")
                            if bnw!="Sì" and bnw!="No" and bnw!="":
                                bnw="X"
                                raise ValueError("\nRisposta non valida.")
                                continue
                            elif bnw=="Sì":
                                os.chdir(PATH)
                                with Image.open(PATH+"\\"+file_name) as image:
                                    image_NEW=Image.new("RGB",(new_width,new_height))
                                    image_NEW=image.convert("L")
                                    image_NEW=image_NEW.resize((new_width,new_height))
                                    if os.path.isdir(file_name.split(".")[0]+"_modified")==False:
                                        os.mkdir(file_name.split(".")[0]+"_modified")
                                    os.chdir(PATH+"\\"+file_name.split(".")[0]+"_modified")
                                    image_NEW.save(root+str(i)+extension)
                                    image_NEW.show()
                                    import matplotlib.pyplot as plt
                                    import numpy as np
                                    image_NEW_array=np.asarray(image_NEW)
                                    plt.imshow(image_NEW_array,cmap="Greys_r")
                                    plt.axis("off")
                                    plt.show()
                            else:
                                os.chdir(PATH)
                                with Image.open(PATH+"\\"+file_name) as image:
                                    image_NEW=Image.new("RGB",(new_width,new_height))
                                    image_NEW=image
                                    image_NEW=image_NEW.resize((new_width,new_height))
                                    if os.path.isdir(file_name.split(".")[0]+"_modified")==False:
                                        os.mkdir(file_name.split(".")[0]+"_modified")
                                    os.chdir(PATH+"\\"+file_name.split(".")[0]+"_modified")
                                    image_NEW.save(root+str(i)+extension)
                                    image_NEW.show()
                                    import matplotlib.pyplot as plt
                                    import numpy as np
                                    image_NEW_array=np.asarray(image_NEW)
                                    plt.imshow(image_NEW_array)
                                    plt.axis("off")
                                    plt.show()
                        except ValueError as e:
                            print(e)
                
                go_on=input("Se si vuole selezionare una nuova immagine da modificare digitare Sì."+
                           "\nSe si vuole modificare nuovamente la stessa immagine digitare No."+
                           "\nSe ci si vuole fermare sigitare Stop:\n")
                i+=1
    except ValueError as e:
        print(e)
    except IndexError:
        print("\nLa directory potrebbe non contenere alcuna immagine.")
        break
```
